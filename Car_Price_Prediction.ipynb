```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df["Fuel_Type"] = le.fit_transform(df["Fuel_Type"])
df["Seller_Type"] = le.fit_transform(df["Seller_Type"])
df ["Transmission"] = le.fit_transform(df["Transmission"])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
column=["Selling_Price","Year","Present_Price","Kms_Driven","Fuel_Type","Seller_Type","Transmission","Owner"]
df[column]=scaler.fit_transform(df[column])
```

```python
sns.heatmap(df.corr(), annot = True)
```

Inference: Year,Present_Price,kms_Driven influence more in determing selling price

```python
X= df.drop(["Selling_Price","Car_Name"],axis=1)
y=df["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=(0.7))
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted
```

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("MAE: ", (metrics.mean_absolute_error(y_predicted, y_test)))
print("MSE: ", (metrics.mean_squared_error(y_predicted, y_test)))
print("R2 score: ", (metrics.r2_score(y_predicted, y_test)))
```
